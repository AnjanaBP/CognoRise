Checking for missing values

```python
df.isnull().sum()
```

Filling the missing values using mean

```python
df['occu_perc']=df['occu_perc'].fillna(df['occu_perc'].mean())
df['capacity']=df['capacity'].fillna(df['capacity'].mean())
df.isna().sum()
```

Changing the data type for date column

```python
df['date']=df['date'].str.replace('-','')
df['date']=df['date'].astype(int)
```

Correlation

```python
df.corr()
```

Visualization of correlation using heatmap

```python
import seaborn as sns
sns.heatmap(df.corr())
```

Dropping unnecessary columns

```python
df.drop(['month','quarter','day'],axis=1)
```

Assigning x and y values

```python
x=df.drop(['total_sales'],axis=1)
x
```

```python
y=df['total_sales']
y
```

Regression plots

```python
import matplotlib.pyplot as plt
sns.regplot(x=df['tickets_sold'],y=y,color='red')
plt.title('Tickets sold Vs Total Sales')
```

```python
sns.regplot(x=df['show_time'],y=y,color='green')
plt.title('Show time Vs Total Sales')
```

```python
sns.regplot(x=df['ticket_use'],y=y,color='blue')
plt.title('Ticket use Vs Total Sales')
```

```python
sns.regplot(x=df['cinema_code'],y=y,color='black')
plt.title('Cinema code Vs Total sales')
```

Feature Selection using Mutual Information Gain

```python
from sklearn.feature_selection import mutual_info_classif
fs=mutual_info_classif(x,y)
fs= pd.Series(fs)
fs.index=x.columns
fs.sort_values(ascending=False)
```

```python
fs.sort_values(ascending=False).plot.bar(figsize=(20, 8))
```

Dropping tickets_out column

```python
df.drop(['tickets_out'],axis=1)
```

Splitting of testing and training data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

Model Creation using Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
df1
```

Performance Evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,r2_score
print("Mean absolute error is",mean_absolute_error(y_test,y_pred))
print("Percentage error is",mean_absolute_percentage_error(y_test,y_pred))
print("R2 Score is",r2_score(y_test,y_pred))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dtree=DecisionTreeRegressor(random_state=42)
dtree.fit(x_train,y_train)
y_pred=dtree.predict(x_test)
y_pred
```

Performance evaluation for decision tree

```python
print("Mean absolute error is",mean_absolute_error(y_test,y_pred))
print("Percentage error is",mean_absolute_percentage_error(y_test,y_pred))
print("R2 Score is",r2_score(y_test,y_pred))
```
